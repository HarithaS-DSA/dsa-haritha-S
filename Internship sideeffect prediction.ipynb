```python
#changing the mistake entries
month_mapping = {
    'Jun': '06',
    'Dec': '12'
}
#converting the age 07-DEC to correct form
data['Age'] = data['Age'].str.replace(r'(\d+)-(\d+)$', r'\1-\2', regex=True)  # Remove hyphen in age groups
data['Age'] = data['Age'].replace(month_mapping, regex=True, inplace=False)
```

# Preprocessing

```python
#dropping the sideeffects column having vacant values

# Replace empty spaces with NaN in the "side effects" column
data['Sides'] = data['Sides'].replace(r'^\s*$', pd.NA, regex=True)

# Drop rows where the "side effects" column contains NaN values
data = data.dropna(subset=['Sides'])

# Reset the DataFrame index
data = data.reset_index(drop=True)
```

```python
# Count vacant spaces in the 'Age' column
vacant_spaces_age = (data['Age'] == ' ').sum()

# Display the count of vacant spaces in the column
print(f"Vacant spaces in 'Age' column: {vacant_spaces_age}")
```

```python
# Count vacant spaces in the 'condition' column
vacant_spaces_condition = (data['Condition'] == ' ').sum()

# Display the count of vacant spaces in the  column
print(f"Vacant spaces in 'Condition' column: {vacant_spaces_condition}")
```

```python
data.replace(' ', np.nan, inplace=True)
```

```python
data.isna().sum()
```

```python
# Group by 'Drug' and fill vacant spaces in 'Condition' with mode
# Define a function to fill vacant spaces in 'Condition' with mode for each 'Drug'
def fill_with_mode(group):
    mode_value = group.mode().iloc[0]
    return group.fillna(mode_value)

# Use the custom function within the transform method
data['Condition'] = data.groupby('Drug')['Condition'].transform(fill_with_mode)

```

```python
#filling null values in age
# Group by "Drug" and find the most frequent age for each drug
most_frequent_ages = data.groupby('Drug')['Age'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Replace NaN values in the "Age" column with the most frequent age for each drug
data['Age'] = data.apply(lambda row: most_frequent_ages[row['Drug']] if pd.isna(row['Age']) else row['Age'], axis=1)
```

```python
#the sideeffects have positive correlation with useful counts,satisfaction,drug,condition,age 

x=data.drop(['Sides','Sex','Reviews','Date','DrugId','EaseofUse','Effectiveness'],axis=1)
y=data['Sides']
```

```python
#scaling the data before modelling

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
#splitting data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
y_pred=dt_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred))
print('confusion_matrix:',confusion_matrix(y_test,y_pred))
```

```python
y_pred
```

```python
print('recall:', recall_score(y_test, y_pred, average=None))
print('f1:', f1_score(y_test, y_pred, average=None))
```

```python
# Get the unique class labels in the target labels
unique_classes = list(set(y_pred))

# Print the number of classes
num_classes = len(unique_classes)
print("Number of classes:", num_classes)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=50, max_depth=10) 

rf_clf.fit(x_train,y_train)
```

```python


# Define the batch size (adjust this value based on your available memory)
batch_size = 1000

# Get the total number of samples in your test data
n_samples = x_test.shape[0]

# Initialize an empty array to store the predicted labels
y_predictions = np.empty(n_samples, dtype=int)

# Iterate in batches
for i in range(0, n_samples, batch_size):
    start = i
    end = min(i + batch_size, n_samples)
    
    # Predict class labels for the current batch
    y_batch = rf_clf.predict(x_test[start:end])
    
    # Store the predicted labels in the corresponding slice of y_predictions
    y_predictions[start:end] = y_batch

# Now, y_predictions contains the predicted class labels for all instances in x_test
```

```python
 y_predictions
```

```python
print('Accuracy_rf:',accuracy_score(y_test,y_predictions))
print('confusion_matrix_rf:',confusion_matrix(y_test,y_predictions))
```

```python
# Get the unique class labels in the target labels
unique_classes = list(set(y_predictions))

# Print the number of classes
num_classes = len(unique_classes)
print("Number of classes:", num_classes)
```
