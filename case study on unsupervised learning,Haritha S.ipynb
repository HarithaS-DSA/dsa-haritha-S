#k means clustering 

# 

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)#kmeans is used to make it uniform
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#from the graph the value is 2
kmeans=KMeans(n_clusters=2,init='k-means++',random_state=42)#kmeans is used to make it uniform
y_kmeans=kmeans.fit_predict(X)
```

```python
labels=kmeans.labels_#get cluster label  for each sample
#add new column to display the cluster
data['clusters']=labels
```

```python
#visualising the clusters formed by applying pca dimensionality reduction
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
#plotting the clusters
plt.scatter(X_pca[:,0], X_pca[:,1],c=y_kmeans)
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.title('K-Means Clustering Result')
plt.show()
```

# silhoutte matrix

```python
silhouette_avg1=silhouette_score(X,y_kmeans)
silhouette_avg1
```

# Aglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))#linkage determine how distance is calculated
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Eucledian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
#plotting the clusters
plt.scatter(X_pca[:,0], X_pca[:,1],c=y_hc)
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.title('Aglomerative Clustering Result')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhoutte_avg=silhouette_score(X,y_hc)
print(silhoutte_avg)
```
